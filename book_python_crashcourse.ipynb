THE CAPITALIZE AND LOWER CASE FUNCTION.

```python
message="There is no shame in repeating the basics, Fahad. In the end you have no one to impress but yourself."
print(message)
print(message.upper()) #EVERY LETTER IS NOW CAPITAL .upper()
print(message.lower()) #every letter is now lower  .lower()
print(message.title()) #Every Letter At The Start Of The Word Is Now Capital .title()
```

NOW LETS SORT THE LISTS IN ALPHABETICAL ORDER

```python
dream_cars=['Porsche 911','Nissan GTR R35','Toyota supra','Acura NSX']
dream_cars.sort()
print(dream_cars)
```

ALPHABETICAL ORDER BUT REVERSE.

```python
dream_cars=['Porsche 911','Nissan GTR R35','Toyota supra','Acura NSX']
dream_cars.sort(reverse=True)
print(dream_cars)
```

SORTING THE LIST TEMPORARILY
