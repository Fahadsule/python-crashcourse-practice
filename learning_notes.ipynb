THE CAPITALIZE AND LOWER CASE FUNCTION.

```python
message="There is no shame in repeating the basics, Fahad. In the end you have no one to impress but yourself."
print(message)
print(message.upper()) #EVERY LETTER IS NOW CAPITAL .upper()
print(message.lower()) #every letter is now lower  .lower()
print(message.title()) #Every Letter At The Start Of The Word Is Now Capital .title()
```

TO REMOVE WHITESPACES ON THE RIGHT.
temporarily

```python
life="Meaningful to the extent of the one living "
print(life.rstrip())    #rstrip()
```

TO REMOVE RIGHT WHITSPACE PERMANENTLY.

```python
life="Meaningful to the extent of the one living "
life=life.rstrip()
print(life)
```

TO REMOVE WHITESPACE TO THE RIGHT

```python
ife="   Meaningful to the extent of the one living"
life=life.lstrip()  #lstrip()
print(life)
```

TO REMOVE ALL UNNECESSARILY WHITE SPACE

```python
name="   NICOLAS MADURO  "
print(name.strip()) #strip()
```
